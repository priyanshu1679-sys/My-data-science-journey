```python
# ==============================
# IMPORT LIBRARIES
# ==============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# NUMPY (Numerical Operations)
# ==============================
print("----- NUMPY -----")

# Create array
arr = np.array([10, 20, 30, 40, 50])
print("Array:", arr)

# Basic operations
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))

# Reshape array
arr2 = np.arange(1, 10).reshape(3, 3)
print("2D Array:\n", arr2)

# ==============================
# PANDAS (Data Handling)
# ==============================
print("\n----- PANDAS -----")

# Create DataFrame
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 90, 78, 92, 88],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

# Display data
print("DataFrame:\n", df)

# Basic operations
print("\nHead:\n", df.head())
print("\nInfo:")
print(df.info())

print("\nDescribe:\n", df.describe())

# Filtering
print("\nStudents with Marks > 85:\n", df[df["Marks"] > 85])

# Add new column
df["Grade"] = ["B", "A", "C", "A", "B"]
print("\nUpdated DataFrame:\n", df)

# ==============================
# MATPLOTLIB (Visualization)
# ==============================
print("\n----- MATPLOTLIB -----")

# Line Plot
plt.figure()
plt.plot(df["Name"], df["Marks"])
plt.title("Marks Line Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# Bar Chart
plt.figure()
plt.bar(df["Name"], df["Marks"])
plt.title("Marks Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(df["Age"], df["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()

# Histogram
plt.figure()
plt.hist(df["Marks"])
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# ==============================
# IMPORT LIBRARIES
# ==============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn style
sns.set()

# ==============================
# NUMPY (Numerical Operations)
# ==============================
print("----- NUMPY -----")

arr = np.array([10, 20, 30, 40, 50])
print("Array:", arr)

print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))

arr2 = np.arange(1, 10).reshape(3, 3)
print("2D Array:\n", arr2)

# ==============================
# PANDAS (Data Handling)
# ==============================
print("\n----- PANDAS -----")

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 90, 78, 92, 88],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

print("DataFrame:\n", df)

print("\nDescribe:\n", df.describe())

# ==============================
# SEABORN VISUALIZATION
# ==============================
print("\n----- SEABORN -----")

# Line Plot
plt.figure()
sns.lineplot(x="Name", y="Marks", data=df)
plt.title("Marks Line Chart")
plt.show()

# Bar Plot
plt.figure()
sns.barplot(x="Name", y="Marks", data=df)
plt.title("Marks Bar Chart")
plt.show()

# Scatter Plot
plt.figure()
sns.scatterplot(x="Age", y="Marks", data=df)
plt.title("Age vs Marks")
plt.show()

# Histogram
plt.figure()
sns.histplot(df["Marks"], kde=True)
plt.title("Marks Distribution")
plt.show()

# Box Plot
plt.figure()
sns.boxplot(x="Marks", data=df)
plt.title("Box Plot of Marks")
plt.show()

# ==============================
# HEATMAP (Correlation)
# ==============================
plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data=[11,14,33,44,55,67,11,23,34,65,64,44,44,65,56,12,23,43,69,1,2,34,45]
plot=['histogram','KERNEL_DENSITY'] # Corrected 'KERNAL_DENSITY' to 'KERNEL_DENSITY'
for plots in plot:
    if plots =='histogram': # Corrected 'Histogram' to 'histogram' for case sensitivity
        sns.histplot(data)
        plt.title("Histogram chart")
        plt.show()
    elif plots =='KERNEL_DENSITY': # Corrected 'KERNAL_DENSITY' to 'KERNEL_DENSITY'
        sns.kdeplot(data)
        plt.title("KERNEL_DENSITY")
        plt.show()

data1=[11,14,33,44,55,67,11,23,34,65,64,44,44,65,56,12,23,43,69,1,2,34,45]
data2=[11,14,3,14,25,67,31,23,64,65,64,84,44,65,56,0,23,43,69,1,2,34,45]
data3=[11,14,3,4,5,67,31,3,64,5,4,84,44,65,56,0,23,43,69,1,2,34,45]
group=[data1,data2,data3]
sns.violinplot(data=group)
plt.title("example of violin plot")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Defining the 'tools' dataframe as it was defined in previous cells
tools = pd.DataFrame({
    'name': ['lulu','shulu','nunu','nuju','ksks','puky','puti','pikachu','tmkc','juju','jushi', 'hello'],
    'age':[17,19,12,32,34,63,65,45,23,56,99, 21],
    'salary':[1,10000,12000,1222,3232,3342,3000,6000,3999,3000,5000, 2100],
    'Team':['ceo','hr','ld','nr','cr','jr','lr','tr', 'qa', 'devops', 'marketing', 'it']
})

# Using numpy to calculate some statistics for labels
total_salary = np.sum(tools['salary'])
avg_age = np.mean(tools['age'])

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# 1. PIE CHART: Salary Distribution by Team
team_salaries = tools.groupby('Team')['salary'].sum()
axes[0].pie(team_salaries, labels=team_salaries.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
axes[0].set_title('Salary Distribution by Team')

# 2. BAR CHART: Age of Employees
axes[1].bar(tools['name'], tools['age'], color='skyblue')
axes[1].axhline(avg_age, color='red', linestyle='--', label=f'Avg Age: {avg_age:.1f}')
axes[1].set_title('Employee Ages')
axes[1].set_xticklabels(tools['name'], rotation=45, ha='right')
axes[1].legend()

# 3. SCATTER PLOT: Age vs Salary
axes[2].scatter(tools['age'], tools['salary'], color='purple', s=100, alpha=0.6)
axes[2].set_title('Age vs Salary Relationship')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Salary')
for i, txt in enumerate(tools['name']):
    axes[2].annotate(txt, (tools['age'].iloc[i], tools['salary'].iloc[i]))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import random

# List of names and roles for variety
names = ['Arjun', 'Sanya', 'Vikram', 'Anjali', 'Rohan', 'Priya', 'Amit', 'Neha', 'Kabir', 'Ishani', 'Rahul', 'Kriti', 'Yash', 'Meera', 'Sahil', 'Tara', 'Dev', 'Riya', 'Aryan', 'Sia', 'Varun', 'Myra', 'Karan', 'Zoya', 'Aditya']
roles = ['Head Chef', 'Sous Chef', 'Waiter', 'Waitress', 'Manager', 'Bartender', 'Cleaner', 'Host']
shifts = ['Morning', 'Evening', 'Night']

# Generating data for 25 employees
data = {
    'Employee_ID': range(101, 126),
    'Name': names,
    'Role': [random.choice(roles) for _ in range(25)],
    'Shift': [random.choice(shifts) for _ in range(25)],
    'Hourly_Wage': [random.randint(15, 35) for _ in range(25)]
}

restaurant_df = pd.DataFrame(data)

print("--- Restaurant Employee Database ---")
display(restaurant_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the college data
college_data = {
    'Year': ['1st Year', '2nd Year', '3rd Year', '4th Year'],
    'Marks': [78, 85, 82, 91]
}

# Converting to DataFrame
college_df = pd.DataFrame(college_data)

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(college_df['Year'], college_df['Marks'], color='teal')
plt.ylim(0, 100) # Setting limit for marks
plt.title('College Marks by Year')
plt.ylabel('Marks (%)')

# Adding values on top of bars
for i, val in enumerate(college_df['Marks']):
    plt.text(i, val + 1, f'{val}%', ha='center', fontweight='bold')
```

```python
# Sample college data: Year vs Marks (Percentage)
college_data = {
    'Year': ['1st Year', '2nd Year', '3rd Year', '4th Year'],
    'Marks': [78, 82, 85, 90]
}

df_college = pd.DataFrame(college_data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df_college['Year'], df_college['Marks'], marker='o', linestyle='-', color='b', linewidth=2, markersize=8)

# Adding titles and labels
plt.title('Academic Progress Over College Years', fontsize=14)
plt.xlabel('Year of Study', fontsize=12)
plt.ylabel('Marks Percentage (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(0, 100) # Setting limit for percentage

# Annotating marks on the points
for i, txt in enumerate(df_college['Marks']):
    plt.annotate(f"{txt}%", (df_college['Year'][i], df_college['Marks'][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```
